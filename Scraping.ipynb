```python
resmenu=[]
m=[]
must_try_final=[]
must_try_list=[]
chef_sp_final=[]
chef_sp_list=[]
links=[]
for i in range(20):
    veg = []
    try:
        link=zomato['Links'][i]
        links.append(link)
        linkpage=requests.get(link, headers=headers)
        print(link)
        soup = BeautifulSoup(linkpage.text, 'lxml')
        sideCol=soup.find_all('p',class_='sc-1hez2tp-0')
        #print(sideCol)
        dc={}
        for i in sideCol:
            val = ""
            if '(' in i.text:
                s1 = len(i.text)
                s=i.text
                for x in range(s1-2,0,-1):
                    if (s[x] == '('):
                        break
                    val = s[x] + val
                dc[((s.split('('))[0]).rstrip()]=int(val)
        resmenu.append(dc)
    
        count=0
        menu={}
        
        VEG = soup.find_all('div',class_='sc-1tx3445-1')
        menuL=soup.find_all('h4',class_='sc-1s0saks-15')
        num = 0
        for k in dc:
            #print(k,dc[k])
            ls=[]
            z = {}
            for j in range(dc[k]):
                #print(count)
                #print(menuL[count].text)
                #print(veg)
                attribute = VEG[num]['type']
                if(attribute == 'veg'):
                    z[menuL[count].text] = 'VEG'
                elif(attribute == 'non-veg'):
                    z[menuL[count].text] = 'NON VEG'
                elif(attribute == 'egg'):
                    z[menuL[count].text] = 'NON VEG'
                else:
                    z[menuL[count].text] = 'NULL'
                #print(veg[num])
                num = num + 1
                count+=1
            menu[k] = z

        m.append(menu)
        
        must_try=soup.find_all('div',class_='sc-1s0saks-17')
        #print(len(must_try))
        must_try_dict={}
        must_try_dishname = {}
        must_try_list=[]
        for x in range(len(must_try)):
            a = must_try[x].find('div',class_ ='sc-2gamf4-0 cRxPpO')
            #print(a)
            v = must_try[x].find('div',class_ ='sc-1tx3445-1')
            item = must_try[x].find('h4',class_ ='sc-1s0saks-15 iSmBPS')
            if(a != None and v['type'] == 'veg'):
                must_try_dishname[item.text] = 'VEG'
            elif(a != None and v['type'] == 'non-veg'):
                must_try_dishname[item.text] = 'NON VEG'
            elif(a != None and v['type'] == 'egg'):
                must_try_dishname[item.text] = 'NON VEG'
        must_try_dict['must_try'] = must_try_dishname
        must_try_final.append(must_try_dict)
        #print(must_try_final)
        
        chef_sp_dict={}
        chef_sp_dishname = {}
        chef_sp_list=[]
        #print(len(chef_sp))
        for y in range(len(must_try)):
            a = must_try[y].find('div',class_ ='sc-2gamf4-0 fQRUpA')
            #print(a)
            v = must_try[y].find('div',class_ ='sc-1tx3445-1')
            item = must_try[y].find('h4',class_ ='sc-1s0saks-15 iSmBPS')
            if(a != None and v['type'] == 'veg'):
                chef_sp_dishname[item.text] = 'VEG'
            elif(a != None and v['type'] == 'non-veg'):
                chef_sp_dishname[item.text] = 'NON VEG'
            elif(a != None and v['type'] == 'egg'):
                chef_sp_dishname[item.text] = 'NON VEG'
        chef_sp_dict['chef_spe']= chef_sp_dishname
        chef_sp_final.append(chef_sp_dict)
        #print(chef_sp_final)
        print("Inserted menu,must,chef",len(m),len(must_try_final),len(chef_sp_final))
    except:
        m.append({})
        must_try_final.append({})
        chef_sp_final.append({})
        print("Inserted menu,must,chef",len(m),len(must_try_list),len(chef_sp_list))
        
```
